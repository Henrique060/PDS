```python
import numpy as np
import matplotlib.pyplot as plt

def ass(f, fase):
    Fs = 1000 
    Ts = 1/Fs
    t = np.arange(0,0.1, Ts)
    x = 9 * np.cos(2*np.pi*f*t+fase)
    plt.figure()
    plt.plot(t,x)
    plt.xlabel('t(s)')
    plt.ylabel('x(t)')
    plt.grid()
    plt.show()

```

```python
from ipywidgets import interact
interact(ass, f=(100,200), fase=(0,np.pi))
```

```python
Fs = 8000
Ts = 1/Fs
t = np.arange(0,1,Ts)
fdelta=5
fc = 400
f1 = fc + fdelta
f2 = fc - fdelta
x1 = 9 * np.cos(2*np.pi*f1*t)
x2 = 9*np.cos(2*np.pi*f2*t)
plt.figure()
y = x1 + x2
plt.plot(t,y)
plt.xlabel('t(s)')
plt.ylabel('x(t)')
from IPython.display import Audio, display
display(Audio(data = y, rate = Fs))
```
