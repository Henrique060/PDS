<h3>1. Usar a função plot para representar sinais contínuos</h3>

<h4> Alínea a) </h4> <br>
$a(t) = 2 \cos(2\pi19t) + \sin(2\pi20t)$

```python
Fs_a = 310 # Frequência de amostragem (> 2*fmax) - escolheu-se este valor, de modo a que o gráfico do sinal se 
           # assemelhasse ao de uma sinusóide
Ts_a = 1/Fs_a
t_a  = np.arange(0, 1, Ts_a) # Criação de um array com os valores de t que se pretendem obter
a    = 2 * np.cos(2 * np.pi * 19 * t_a) + 5 * np.sin(2 * np.pi * 20 * t_a) # Obtenção dos valores respetivos a cada t

plt.title('1. Gráfico do Sinal a)') # Título
plt.xlabel('t(s)')                  # Variável no eixo das coordenadas
plt.ylabel('a(t)')                  # Variável no eixo das ordenadas
plt.plot(t_a, a)                    # Gráfico
plt.grid()
```

<h4> Alínea b) </h4> <br>
$b(t) = (1 + \frac{1}{2} \cos(2\pi12t)) \cos(2\pi200t)$

```python
Fs_b = 2000
Ts_b = 1/Fs_b
t_b  = np.arange(0, 0.15, Ts_b)
b    = (1 + (1/2) * np.cos(2 * np.pi * 12 * t_b)) * np.cos(2 * np.pi * 200 * t_b)

plt.title('1. Gráfico do Sinal b)') 
plt.xlabel('t(s)') 
plt.ylabel('b(t)') 
plt.plot(t_b, b)
plt.grid()
```

<h4> Alínea c) </h4> <br>
$c(t) = e^{-20t^2} \cos(2\pi100t) (u(t + 3) - u(t - 3))$

```python
Fs_c = 400
Ts_c = 1/Fs_c
t_c = np.arange(-0.4, 0.4, Ts_c)

# u é uma função cujos valores são iguais a 0 para coordenadas negativas e a 1 para coordenadas positivas
def u(num):
    u = np.zeros(t_c.shape)
    u[num > 0.0] = 1.0
    u[num < 0.0] = 0.0
    return u

c = pow(np.e, -20 * t_c ** 2) * np.cos(2 * np.pi * 100 * t_c) * (u(t_c + 3) - u(t_c - 3))

plt.xlabel('t(s)')
plt.ylabel('c(t)')
plt.title('1. Gráfico do Sinal c)')
plt.plot(t_c, c)
plt.grid()
```

```python
# Aproximação do sinal
t_c = np.arange(0, 0.4, Ts_c)
c = pow(np.e, -20 * t_c ** 2) * np.cos(2 * np.pi * 100 * t_c) * (u(t_c + 3) - u(t_c - 3))
plt.plot(t_c, c)
plt.xlabel('t(s)')
plt.ylabel('c(t)')
plt.title('1. Aproximação do Gráfico do Sinal c)')
plt.grid()
```

<h4> Alínea d) </h4> <br>
$d(t) = \frac{4}{\pi} \sum_{k = 1}^N \frac{\sin(2\pi(2k - 1)f_{0} t)}{2k -1}$
<br><br>
$N \in $ {1, 10, 1000}
<br><br>
Para a realização deste exercício, representar-se-ão 3 gráficos diferentes, nomeados como $di)$, $dii)$ e $diii)$, correspondentes a cada somatório.

$di)$

```python
# Para N = 1
f0_di = 5
Fs_di = 200
Ts_di = 1/Fs_di
t_di  = np.arange(0, 0.4, Ts_di)
di    = np.zeros(len(t_di))

for k in range(1, 1, 1):
    di += ((np.sin(2 * np.pi * (2 * k - 1) * f0_di * t_di))/(2 * k - 1)) 

plt.plot(t_di, (4 / np.pi) * di)
plt.title('1. Gráfico do Sinal di)')
plt.xlabel('t(s)')
plt.ylabel('di(t)')
plt.grid()
```

<h3>4. Desenvolver uma função que crie composições musicais baseadas em sinusóides.</h3>

```python
Fs = 8000

melodia = [('E', 2), ('P', 0.3), ('E', 1.25), ('B', 0.25), ('A', 1), ('G', 1), ('F#', 1), ('E', 1), 
           ('P', 0.3), ('E', 1.25), ('D', 0.5), ('B', 0.25), ('C#', 1), ('A', 1), ('B', 1)]

dicionarioNotas = {'E':39, 'F#':41, 'G':42, 'A':44, 'B':46, 'C':47, 'C#':48, 'D':49, 'P':0}
teclas = np.array([39, 41, 42, 44, 46, 47, 48, 49, 0])

def musica(melodia):
    musica = []

    for i in range(len(melodia)):
        nota = melodia[i][0]
        tempo = melodia[i][1]
        t = np.arange(0, tempo, 1/Fs)
        tecla = dicionarioNotas[nota]
        freq = (2 ** ((tecla - 49)/12)) * 440
        x = np.cos(2 * np.pi * freq * t)
        musica.extend(x)
    return musica
    
display(Audio(data = musica(melodia), rate = Fs))
```

<h4> Desenhar a etapa (com polígonos) </h4>

```python
m = Map(center=[43.05825255802987, -0.024719238281250003], zoom=10)

multi_route = Polyline(locations = [ marco['latlng'], maxime['latlng'], michal['latlng'],
                                   steven['latlng'], thibaut['latlng'], thomas['latlng']],
                                    color="green", fill = False)

m.add_layer(multi_route)
display(m)

## tendo em conta que a rota é sempre a mesma, verificamos que os percursos sobrepõem-se, havendo ligeiras variações
## devido a erros de medição, ligeiros desvios, etc
```

<h4> Calcular o número de kms a subir </h4>

```python
nomes = ['Marco', 'Maxime', 'Michal', 'Steven', 'Thibaut', 'Thomas']
a_subir = np.zeros(len(ciclistas))

for c in range(len(ciclistas)):
    subir = 0
    acumulada = 0
    for l in range(len(ciclistas[c]['grade_smooth'])):
        if l != 0:
            acumulada = ciclistas[c]['distance'][l-1]
            
        if (ciclistas[c]['grade_smooth'][l] > 0): #se estiver a subir
            subir += ciclistas[c]['distance'][l] - acumulada #adicionamos os metros percorridos
        
    a_subir[c] = subir/1000 #converter em kms
    

plt.grid()
plt.title('Km a subir')
plt.xlabel('Ciclistas')
plt.ylabel('Kilómetros')

plt.stem(nomes, a_subir, use_line_collection = True)
a_subir
```

<h4> Calcular o número de kms a descer </h4>

```python
nomes = ['Marco', 'Maxime', 'Michal', 'Steven', 'Thibaut', 'Thomas']
a_descer = np.zeros(len(ciclistas))

for c in range(len(ciclistas)):
    descer = 0
    acumulada = 0
    for l in range(len(ciclistas[c]['grade_smooth'])):
        if l != 0:
            acumulada = ciclistas[c]['distance'][l-1]
            
        if (ciclistas[c]['grade_smooth'][l] < 0): #se estiver a descer
            descer += ciclistas[c]['distance'][l] - acumulada #soma-se a distância percorrida
        
    a_descer[c] = descer/1000 #converter em km
    

plt.grid()
plt.title('Km a descer')
plt.xlabel('Ciclistas')
plt.ylabel('Kilómetros')

plt.stem(nomes, a_descer, use_line_collection = True)
a_descer
        
```

<h4> Determinar a velocidade média em todo o percurso</h4>

```python
vel_media = np.zeros(len(ciclistas))

for c in range(len(ciclistas)):
    vel_total = 0
    
    if not 'velocity_smooth' in ciclistas[c].keys(): #se não existem dados calculamos a vel media com o que temos
        vel_media[c] = ciclistas[c]['distance'][-1] / ciclistas[c]['time'][-1] # vel = distancia/tempo
    else:
        for v in range(len(ciclistas[c]['velocity_smooth'])):
            vel_total += ciclistas[c]['velocity_smooth'][v]
        vel_media[c] = vel_total / len(ciclistas[c]['velocity_smooth'])
        

plt.title('Velocidade Média (m/s)')
plt.xlabel('Ciclistas')
plt.ylabel('m/s')
plt.grid()
plt.stem(nomes, vel_media, use_line_collection = True)
vel_media
```

<h4> Determinar a velocidade média nos troços a descer</h4>

```python
a_descer = np.zeros(len(ciclistas))
media_descer = np.zeros(len(ciclistas))

for c in range(len(ciclistas)):
    descer = 0
    tempo_descer = 0
    acumulada = 0
    
    for l in range(len(ciclistas[c]['grade_smooth'])):
        if l != 0:
            acumulada = ciclistas[c]['distance'][l-1]
        
        if (ciclistas[c]['grade_smooth'][l] < 0):
            descer += ciclistas[c]['distance'][l] - acumulada
            tempo_descer += 1
            
    a_descer[c] = descer
    media_descer[c] = a_descer[c] / tempo_descer
    
plt.grid()
plt.title('Velocidade Média em Descidas (m/s)')
plt.xlabel('Ciclistas')
plt.ylabel('m/s')
plt.stem(nomes, media_descer, use_line_collection = True)
media_descer
```

<h4> Comparar a evolução do participante que ficou melhor colocado com a do que ficou pior colocado </h4>
