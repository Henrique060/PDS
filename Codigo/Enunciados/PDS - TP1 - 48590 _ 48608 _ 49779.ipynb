<h3>1. Usar a função plot para representar sinais contínuos</h3>

Alínea a)

```python
import numpy as np
import matplotlib.pyplot as plt 

Fs = 8000

t = np.arange(0, 1, 1/Fs)
a = 2*np.cos(2*np.pi*19*t)+5*np.sin(2*np.pi*20*t)

plt.title('1. Gráfico do Sinal a)') # Título
plt.xlabel('t(s)') # Variável no eixo das coordenadas
plt.ylabel('a(t)') # Variável no eixo das ordenadas
plt.plot(t,a)
```

Alínea b)

```python
Fs= 1000
t = np.arange(0, 1, 1/Fs)
b = (1+1/2*np.cos(2*np.pi*12*t))*np.cos(2*np.pi*200*t)
plt.title('1. Gráfico do Sinal b)') # Título
plt.xlabel('t(s)') # Variável no eixo das coordenadas
plt.ylabel('b(t)') # Variável no eixo das ordenadas
plt.plot(t, b)
```

Alínea c)

```python
Fs = 1000
t = np.arange(-0.6, 0.6, 1/Fs)

# u é uma função cujos valores são iguais a 0 para coordenadas negativas, e 1 para coordenadas positivas
def u(num):
    u = np.zeros(t.shape)
    u[num > 0.0] = 1.0
    u[num < 0.0] = 0.0
    return u

c = pow(np.e,-20*t**2) * np.cos(2*np.pi*100*t) * (u(t + 3) - u(t - 3))

plt.title('1. Gráfico do Sinal c)') # Título
plt.xlabel('t(s)') # Variável no eixo das coordenadas
plt.ylabel('c(t)') # Variável no eixo das ordenadas
plt.plot(t,c)
```

<h3>2. Analisar os sinais no domínio da frequência, representando os espectros de amplitude, usando a fft.</h3>

Alínea a)

```python
A = np.fft.fft(a)
N_A = len(a)
Fs_A = 1000

freq = np.arange(N_A) * Fs_A/N_A
plt.stem(freq, np.abs(A))
plt.plot(freq, np.abs(A))
```

Alínea b)

```python
B = np.fft.fft(b)
N_B = len(b)
Fs_B = 1000

freq = np.arange(N_B) * Fs_B/N_B
plt.stem(freq, np.abs(B))
plt.plot(freq, np.abs(B))
```
