<h3>1. Usar a função plot para representar sinais contínuos</h3>

<h4> Alínea a) </h4> <br>
$a(t) = 2 \cos(2\pi19t) + \sin(2\pi20t)$

```python
Fs_a = 310 # Frequência de amostragem (> 2*fmax) - escolheu-se este valor, de modo a que o gráfico do sinal se 
           # assemelhasse ao de uma sinusóide
Ts_a = 1/Fs_a
t_a  = np.arange(0, 1, Ts_a) # Criação de um array com os valores de t que se pretendem obter
a    = 2 * np.cos(2 * np.pi * 19 * t_a) + 5 * np.sin(2 * np.pi * 20 * t_a) # Obtenção dos valores respetivos a cada t

plt.title('1. Gráfico do Sinal a)') # Título
plt.xlabel('t(s)')                  # Variável no eixo das coordenadas
plt.ylabel('a(t)')                  # Variável no eixo das ordenadas
plt.plot(t_a, a)                    # Gráfico
plt.grid()
```

<h4> Alínea b) </h4> <br>
$b(t) = (1 + \frac{1}{2} \cos(2\pi12t)) \cos(2\pi200t)$

```python
Fs_b = 2000
Ts_b = 1/Fs_b
t_b  = np.arange(0, 0.15, Ts_b)
b    = (1 + (1/2) * np.cos(2 * np.pi * 12 * t_b)) * np.cos(2 * np.pi * 200 * t_b)

plt.title('1. Gráfico do Sinal b)') 
plt.xlabel('t(s)') 
plt.ylabel('b(t)') 
plt.plot(t_b, b)
plt.grid()
```

<h4> Alínea c) </h4> <br>
$c(t) = e^{-20t^2} \cos(2\pi100t) (u(t + 3) - u(t - 3))$

```python
Fs_c = 400
Ts_c = 1/Fs_c
t_c = np.arange(-0.4, 0.4, Ts_c)

# u é uma função cujos valores são iguais a 0 para coordenadas negativas e a 1 para coordenadas positivas
def u(num):
    u = np.zeros(t_c.shape)
    u[num > 0.0] = 1.0
    u[num < 0.0] = 0.0
    return u

c = pow(np.e, -20 * t_c ** 2) * np.cos(2 * np.pi * 100 * t_c) * (u(t_c + 3) - u(t_c - 3))

plt.xlabel('t(s)')
plt.ylabel('c(t)')
plt.title('1. Gráfico do Sinal c)')
plt.plot(t_c, c)
plt.grid()
```

```python
# Aproximação do sinal
t_c = np.arange(0, 0.4, Ts_c)
c = pow(np.e, -20 * t_c ** 2) * np.cos(2 * np.pi * 100 * t_c) * (u(t_c + 3) - u(t_c - 3))
plt.plot(t_c, c)
plt.xlabel('t(s)')
plt.ylabel('c(t)')
plt.title('1. Aproximação do Gráfico do Sinal c)')
plt.grid()
```

<h4> Alínea d) </h4> <br>
$d(t) = \frac{4}{\pi} \sum_{k = 1}^N \frac{\sin(2\pi(2k - 1)f_{0} t)}{2k -1}$
<br><br>
$N \in $ {1, 10, 1000}
<br><br>
Para a realização deste exercício, representar-se-ão 3 gráficos diferentes, nomeados como $di)$, $dii)$ e $diii)$, correspondentes a cada somatório.

$di)$

```python
# Para N = 1
f0_di = 5
Fs_di = 200
Ts_di = 1/Fs_di
t_di  = np.arange(0, 0.4, Ts_di)
di    = np.zeros(len(t_di))

for k in range(1, 1, 1):
    di += ((np.sin(2 * np.pi * (2 * k - 1) * f0_di * t_di))/(2 * k - 1)) 

plt.plot(t_di, (4 / np.pi) * di)
plt.title('1. Gráfico do Sinal di)')
plt.xlabel('t(s)')
plt.ylabel('di(t)')
plt.grid()
```

$dii)$

```python
# Para N = 10
f0_dii = 5
Fs_dii = 800
Ts_dii = 1/Fs_dii
t_dii  = np.arange(0, 0.4, Ts_dii)
dii    = np.zeros(len(t_dii))

for k in range(1, 10, 1):
    dii += ((np.sin(2 * np.pi * (2 * k - 1) * f0_dii * t_dii))/(2 * k - 1)) 

plt.plot(t_dii, (4 / np.pi) * dii)
plt.title('1. Gráfico do Sinal dii)')
plt.xlabel('t(s)')
plt.ylabel('dii(t)')
plt.grid()
```

$diii)$

```python
# Para N = 1000
f0_diii = 5
Fs_diii = 900
Ts_diii = 1/Fs_diii
t_diii  = np.arange(0, 0.4, Ts_diii)
diii    = np.zeros(len(t_diii))

for k in range(1, 1000, 1):
    diii += ((np.sin(2 * np.pi * (2 * k - 1) * f0_diii * t_diii))/(2 * k - 1)) 

plt.plot(t_diii, (4 / np.pi) * diii)
plt.title('1. Gráfico do Sinal diii)')
plt.xlabel('t(s)')
plt.ylabel('diii(t)')
plt.grid()
```

<h4> Alínea e) </h4> <br>
$e(t) = \frac{8}{\pi^2} \sum_{k = 1}^N (-1)^k (\frac{\sin(2\pi(2k - 1)f_{0} t)}{(2k-1)^2})$
<br><br>
$N \in $ {1, 10, 1000}
<br><br>
Para a realização deste exercício, representar-se-ão 3 gráficos diferentes, nomeados como $ei)$, $eii)$ e $eiii)$, correspondentes a cada somatório.

$ei)$

```python
# Para N = 1
f0_ei = 1
Fs_ei = 10
Ts_ei = 1/Fs_ei
t_ei  = np.arange(-0.5, 1.5, Ts_ei)
ei    = np.zeros(len(t_ei))

for k in range(1, 1, 1):
    ei += ((-1) ** k) * ((np.sin(2 * np.pi * (2 * k - 1) * f0_ei * t_ei))/(pow(2 * k - 1, 2)))

plt.plot(t_ei, (8 / np.pi ** 2) * ei)
plt.title('1. Gráfico do Sinal ei)')
plt.xlabel('t(s)')
plt.ylabel('ei(t)')
plt.grid()
```

```python
# Para N = 10
f0_eii = 1
Fs_eii = 100
Ts_eii = 1/Fs_eii
t_eii  = np.arange(-0.5, 1.5, Ts_eii)
eii    = np.zeros(len(t_eii))

for k in range(1, 10, 1):
    eii += ((-1) ** k) * ((np.sin(2 * np.pi * (2 * k - 1) * f0_eii * t_eii))/(pow(2 * k - 1, 2)))

plt.plot(t_eii, (8 / np.pi ** 2) * eii)
plt.title('1. Gráfico do Sinal eii)')
plt.xlabel('t(s)')
plt.ylabel('eii(t)')
plt.grid()
```

$eiii)$

```python
# Para N = 1000
f0_eiii = 1
Fs_eiii = 60
Ts_eiii = 1/Fs_eiii
t_eiii  = np.arange(-0.5, 1.5, Ts_eiii)
eiii    = np.zeros(len(t_eiii))

for k in range(1, 1000, 1):
    eiii += ((-1) ** k) * ((np.sin(2 * np.pi * (2 * k - 1) * f0_eiii * t_eiii))/(pow(2 * k - 1, 2)))

plt.plot(t_eiii, (8 / np.pi ** 2) * eiii)
plt.title('1. Gráfico do Sinal eiii)')
plt.xlabel('t(s)')
plt.ylabel('eiii(t)')
plt.grid()
```

<h3>4. Desenvolver uma função que crie composições musicais baseadas em sinusóides.</h3>

```python
Fs = 8000

melodia = [('E', 2), ('P', 0.3), ('E', 1.25), ('B', 0.25), ('A', 1), ('G', 1), ('F#', 1), ('E', 1), 
           ('P', 0.3), ('E', 1.25), ('D', 0.5), ('B', 0.25), ('C#', 1), ('A', 1), ('B', 1)]

dicionarioNotas = {'E':39, 'F#':41, 'G':42, 'A':44, 'B':46, 'C':47, 'C#':48, 'D':49, 'P':0}
teclas = np.array([39, 41, 42, 44, 46, 47, 48, 49, 0])

def musica(melodia):
    musica = []

    for i in range(len(melodia)):
        nota = melodia[i][0]
        tempo = melodia[i][1]
        t = np.arange(0, tempo, 1/Fs)
        tecla = dicionarioNotas[nota]
        freq = (2 ** ((tecla - 49)/12)) * 440
        x = np.cos(2 * np.pi * freq * t)
        musica.extend(x)
    return musica
    
display(Audio(data = musica(melodia), rate = Fs))
```
