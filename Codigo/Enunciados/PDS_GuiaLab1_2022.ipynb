## Exercicio 1 - Sinusoide
Seja $x(t)= 9 \cos(2\pi50t + 0.4\pi)$
Defina a função numa base de tempo adequada.

```python

import numpy as np
import matplotlib.pyplot as plt
t = np.arange(-0.05, 0.05, 0.0001)
x = 9 * np.cos(2 * np.pi * 50 * t + 0.4 * np.pi)

plt.plot(t,x )
```

A partir da análise do gráfico determine os valores da amplitude ($A$), fase ($\phi$) e frequencia linear ($f$) associada à representação:

$ x(t) = A \cos(2\pi f t + \phi)$

Gere o código para implementar esta sinusóide

```python
# A = 8
# T = 12.5
# f = 1/12.5
# fase = 2 * np.pi * 1/T * desfasamento (neste caso 2.5)

import numpy as np
import matplotlib.pyplot as plt

t = np.arange(-15,20,0.01)
fase = 2* np.pi * 2.5/12.5 # 1/12.5 = f * t1
x = 8 * np.cos(2 * np.pi * 0.08 * t - fase)

plt.grid()
plt.plot(t,x)
```

## Exercicio 3 - Soma de Sinusoides

Calcule analiticamente a soma das duas sinusóides.

$x(t)=2 \cos⁡(2\pi 10t+\pi/4)+\cos(2 \pi 10t-\pi /3)$, para t $\in$ [-1,3] 

Use a função plot para representar os sinal no intervalo de tempo indicado.

```python
import numpy as np 
import matplotlib.pyplot as plt

t = np.arange(-1,3,0.01)

x = 2 * np.cos(2 * np.pi * 10 * t + np.pi/4)
y = np.cos(2 * np.pi * 10 * t - np.pi/3)

z = x + y

plt.grid()
plt.plot(t,z)
```

## Exercicio 4 - Geração de Sinais Discretos

Use a função stem para representar os sinais discretos   nos intervalos indicados.
* $x[n] = \cos[2\pi n/16]$, n $\in$ [0,32]
* $x[n] = \frac{10}{\pi n}\cos[\pi n/10]$, n $\in$ [-50,50]

```python
import matplotlib.pyplot as plt
import numpy as np
n = np.arange(32)
x1= np.cos(2*np.pi*n/16)
plt.stem(n,x1)
#plt.plot(n,x1,'*-')
x2 = (10 / (np.pi * n)) * np.cos(np.pi * n/10) #dá erro devido a encontrar infinitos e zeros
n2 = np.arange(-50,50)
plt.figure()
plt.stem(n2,x2)
```
