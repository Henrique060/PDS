```python
##TESTE 2021 EX3(B)


import numpy as np
import matplotlib.pyplot as plt

Fs = 1000
t = np.arange(-3,5,1/Fs)
f0 = 1/4
kmax = 30
ks = np.arange(-kmax, kmax, 1)

x = np.zeros(len(t))

for i in range(len(ks)):
    if ks[i] == 0:
        x = x + 0
    else:
        x = x + (-4)/(1j * ks[i] * np.pi) * np.exp(-1j * 2 * np.pi * ks[i]/4) * np.exp(1j * 2 * np.pi * ks[i] * f0 * t) ## a segunda exponencial a ser multiplicada é a parte imaginaria da equação sinusoidal (ex: cos(x) + sen(x))

plt.plot(t,x)



##nota potencia do sinal
## Px = np.sum(.....)/len(x) se for preciso no dominio da frequencia
## em caso de ser so a soma das amplitudes como normal nao se dvidie por len(x)
```
