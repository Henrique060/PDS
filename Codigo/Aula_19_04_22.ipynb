```python

from IPython.display import Audio, display
import numpy as np
import matplotlib.pyplot as plt

Fs = 8000
notas=[('A')]
dicionarioNotas = {'C':35, 'D':37, 'E': 39, 'F':40, 'G':42, 'A':44, 'B':46}
teclas = np.array([35,37,39,40,42,44,46])
freqs = 2**((teclas-49)/12) * 440
print(freqs)

tecla = dicionarioNotas['C']
freq = 2**((tecla-49)/12) * 440
#freq


melodia = [('E',1),('E',1),('F',1),('G',1),('G',1),('F',1),('E',1),('D',1),('C',1),('C',1),('D',1),('E',1),('E',1),('E',0.5),('D',1),('D',1)]
mel = [('G',0.6),('A',0.6),('G',0.6),('F',0.6),('D',1.05),('G',0.6),('A',0.6),('G',0.6),('F',0.8),('D',2.1), ('C',1.5),('D',0.6),
       ('F',0.6), ('G',0.9),('A',1.2), ('D',4.2)]
musica=[]
for i in range(len(mel)):
    nota = mel[i][0]
    tecla=dicionarioNotas[nota]
    freq = 2**((tecla-49)/12)*440
    print(freq)
    t=np.arange(0,mel[i][1],1/Fs)
    x = np.cos(2*np.pi*freq*t)
    musica.extend(x)
display(Audio(data=musica, rate = Fs*3))

print(len(musica)/8000)


#espectograma

plt.specgram(musica, Fs = Fs)
plt.colorbar()
```
